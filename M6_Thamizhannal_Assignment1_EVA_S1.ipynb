```
# Name: Thamizhannal
# Date: 28-Apr-2019
#
# Import numpy - to perform convolution and math operations between imput image with kernel
# Import urlli - to load image from url
# Import cv2 - open cv package for image manipulation
# import matplotlib - to plot image in screen

import numpy as np
from urllib.request import urlopen
import cv2
from matplotlib import pyplot as plt

# METHOD #1: OpenCV, NumPy, and urllib
def url_to_image(url):
	# download the image, convert it to a NumPy array, and then read
	# it into OpenCV format
	resp = urlopen(url)
	image = np.asarray(bytearray(resp.read()), dtype="uint8")
	image = cv2.imdecode(image, cv2.IMREAD_COLOR)
 
	# return the image
	return image
```

```
# Dowload helvetica image url and convert it to numpy image
image = url_to_image("https://fontsarena-cd5e.kxcdn.com/wp-content/uploads/2019/04/helvetica-now-font-400x364.png")
```

```
from google.colab.patches import cv2_imshow
# Display image in colab
cv2_imshow(image)
```

```
# Canny edge detection in opencv
edges = cv2.Canny(image,100,200)
```

### Kernel
In computer vision, kernel/filter is a <b> 3x3</b> numerical matrix that essentially used to perform operations such as <b> horizontal/vertical edge detection, blurring, sharpening</b> in the given input image. This is achieved by convolving input image with kernel that has specific numerical values according to the operation chosen.

### Convolution 
convolution is a <b> matrix multiplication and summation operation </b> applied between input data image matrix with kernel (data matrix) by moving kernel on image from top left corner of image to bottom right corner of an image by 1 pixel at a time towards right side. 2D input image data matrix of size 5x5 convolved with kernel of size 3x3 outputs a matrix of size 3x3 is so called feature matrix(map)


### Vertical Edge detector

```
kernel = np.float32([[-1,0,1],[-1,0,1],[-1,0,1]])
dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

### Horizontal Edge Detector

```
# Horizondal edge detector
kernel = np.float32([[1,1,1],[0,0,0],[-1,-1,-1]])
dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

### 45 Degree Angle Detector

```
kernel = np.float32([[2, -1, -1],[-1,2,-1],[-1,-1,2]])
dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

###  Sharpen Kernel

```
kernel = np.float32([[0,-1,0],[-1,5,-1],[0,-1,0]])
dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

### Blue Kernel

```

kernel = np.float32([[0.0625,0.125,0.0625],[0.125,0.25,0.125],[0.0625,0.125,0.0625]])
dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

### Identity detector

```
kernel = np.float32([[0,0,0],[0,1,0],[0,0,0]])
dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)

```
